### Ejercicios APE1

Calcular la conductividad térmica de un producto alimenticio desarrollado para suplir necesidades energéticas, el cual (alimento) tiene una composición de 90% de agua, 7% de carbohidratos, 1.5% de proteína, 0.5 de cenizas y 0 % de grasa.

```python
Xc= 7/100
Xp= 1.5/100
Xf= 0/100
Xa= 0.5/100
Xw= 90/100
K= ( 0.25*Xc + 0.155*Xp + 0.16*Xf + 0.135*Xa + 0.58*Xw)
print("El valor de la conductividad térmica del producto es",K)
```

Calcular la conductividad térmica de pure de vegateles con 92% de agua.

```python
Xw= 92/100
K= 0.148 + 0.493*Xw
print("El valor de la conductividad térmica del pures es",K)
```

Calcular la conductividad térmica de un alimento líquido a base de leche, si la temperatura (T) = 25°C

```python
T= 25
Xw= 98/100
K= ( 326.8 + 1.0413*T - 0.00337*T**2)*(0.44 + 0.54*Xw)*1.73e-3
print("El valor de la conduvtividad térmica es",K)
```
